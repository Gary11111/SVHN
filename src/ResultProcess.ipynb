```python
import pickle
import numpy as np
from prettytable import PrettyTable as pt
import matplotlib.pyplot as plt



def analyse(arrLoss, arrAcc,name):
    print('max loss of %s = %f' % (name,max(arrLoss)))
    print('min loss of %s = %f' % (name,min(arrLoss)))
    print('mean loss of %s = %f'% (name,np.mean(arrLoss)))
    print('std loss of %s = %f' % (name,np.std(arrLoss)))
    print('max acc of %s = %f' % (name,max(arrAcc)))
    print('min acc of %s = %f' % (name,min(arrAcc)))
    print('mean loss of %s = %f'% (name,np.mean(arrAcc)))
    print('std loss of %s = %f' % (name,np.std(arrAcc)))
    
    return [max(arrLoss),min(arrLoss),np.mean(arrLoss),np.std(arrLoss)],[max(arrAcc),min(arrAcc),np.mean(arrAcc),np.std(arrAcc)]
```
